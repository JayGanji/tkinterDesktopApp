```python
from tkinter import *
from PIL import Image,ImageTk
import random

root = Tk()
root.title('Final Project - Semester Viewer - Desktop GUI')
root.configure(background="black")
root.iconbitmap('rvcc_icon.ico')

headerLogo = ImageTk.PhotoImage(Image.open('rvcc_logo.jpg'))
logoLabel = Label(root, image=headerLogo)
logoLabel.grid(row=0, column=1, sticky=N, pady=(10,0))

headerLabel = Label(root, text="Welcome to the Python Course Apps Menu", fg="white", bg="black", font="none 14 bold")
headerLabel.grid(row=1, column=1, padx=20, pady=(20,0))
instructions = Label(root, text="Click a button to open the subprogram in a new window", fg="white", bg="black", font="none 10")
instructions.grid(row=2, column=1)



def bitcoinConverter():
    newWindow = Toplevel()
    newWindow.title('Bitcoin Coverter Program')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info = Label(newWindow, text="The current trade value to USD for 1 btc is $48,425.04", fg="#82b843", bg="black", font="none 12")
    info.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
    instructions = Label(newWindow, text="Enter the number of bitcoins you have:", fg="white", bg="black")
    instructions.grid(row=1, column=0, sticky=E, padx=(10,0), pady=10)
    
    userInputBox = Entry(newWindow, width=25)
    userInputBox.grid(row=1, column=1, padx=(0,10), pady=10)

    
    def calculate():
        
        userInput = float(userInputBox.get())
    
        btcUsdValue = 48425.04
        userUsdValue = round(userInput * btcUsdValue, 2)
        answer = '$'+str(userUsdValue)
        
        answerLabel = Label(newWindow, text=answer, width=25, fg="white", bg="black")
        answerLabel.grid(row=3, column=0, columnspan=2, padx=10, pady=10)
        
        userInputBox.delete(0,END)
    
    calculateButton = Button(newWindow, text="Calculate", command=calculate)
    calculateButton.grid(row=2, column=0, columnspan=2, padx=10, pady=10)
    

    quitAppButton = Button(newWindow, text="Exit Subprogram", command=newWindow.destroy, fg="black", bg="red").grid(row=5, column=1, padx=(0, 10), pady=(0, 10),sticky=E)



    
def nonsensicalStory():
    newWindow = Toplevel()
    newWindow.title('Nonsensical Story Program')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info = Label(newWindow, text="Fill in the words for a nonsensical story", fg="#82b843", bg="black", font="none 12")
    info.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
    
    inputName = Label(newWindow, text="Enter a given name:", fg="white", bg="black").grid(row=1, column=0, sticky=E, padx=(10,0), pady=5)
    nameInputBox = Entry(newWindow, width=25)
    nameInputBox.grid(row=1, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputLocation = Label(newWindow, text="Enter a location:", fg="white", bg="black").grid(row=2, column=0, sticky=E, padx=(10,0), pady=5)
    locationInputBox = Entry(newWindow, width=25)
    locationInputBox.grid(row=2, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputAdj1 = Label(newWindow, text="Enter an adjective:", fg="white", bg="black").grid(row=3, column=0, sticky=E, padx=(10,0), pady=5)
    adj1InputBox = Entry(newWindow, width=25)
    adj1InputBox.grid(row=3, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputNoun = Label(newWindow, text="Enter a noun:", fg="white", bg="black").grid(row=4, column=0, sticky=E, padx=(10,0), pady=5)
    nounInputBox = Entry(newWindow, width=25)
    nounInputBox.grid(row=4, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputAdj2 = Label(newWindow, text="Enter another adjective:", fg="white", bg="black").grid(row=5, column=0, sticky=E, padx=(10,0), pady=5)
    adj2InputBox = Entry(newWindow, width=25)
    adj2InputBox.grid(row=5, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputVerbP = Label(newWindow, text="Enter a verb in the past tense:", fg="white", bg="black").grid(row=6, column=0, sticky=E, padx=(10,0), pady=5)
    verbPInputBox = Entry(newWindow, width=25)
    verbPInputBox.grid(row=6, column=1, sticky=W, padx=(0,10), pady=5)
    
    def createStory():
        storyPhrase1 = (nameInputBox.get() + " was at the " + locationInputBox.get() + " and they felt " + 
                       adj1InputBox.get() + ". ")
        storyPhrase2 = (nameInputBox.get() + " asked the first person they saw there for a " 
                       + nounInputBox.get() + ". They magically conjured it out of thin air.")
        storyPhrase3 = ("The problem was, it was " + adj2InputBox.get() + " so " + nameInputBox.get() + " just " 
                        + verbPInputBox.get() + " instead for the day.")

        
        storyLabel1 = Label(newWindow, text=storyPhrase1, width=125,fg="#82b843", bg="black")
        storyLabel1.grid(row=8, column=0, columnspan=2, padx=10, pady=(20,0))
        storyLabel2 = Label(newWindow, text=storyPhrase2, width=125,fg="#82b843", bg="black")
        storyLabel2.grid(row=9, column=0, columnspan=2, padx=10, pady=0)
        storyLabel3 = Label(newWindow, text=storyPhrase3, width=125,fg="#82b843", bg="black")
        storyLabel3.grid(row=10, column=0, columnspan=2, padx=10, pady=(0,20))
        
        nameInputBox.delete(0,END)
        locationInputBox.delete(0,END)
        adj1InputBox.delete(0,END)
        nounInputBox.delete(0,END)
        adj2InputBox.delete(0,END)
        verbPInputBox.delete(0,END)
    
    createStoryButton = Button(newWindow, text="Create Story", command=createStory)
    createStoryButton.grid(row=7, column=0, columnspan=2, padx=10, pady=10)
    
    
    

    quitAppButton = Button(newWindow, text="Exit Subprogram", command=newWindow.destroy, fg="black", bg="red").grid(row=11, column=1, padx=(0, 10), pady=(0, 10),sticky=E)    
    
    
def artwork():
    """function to display some turtle artwork"""
    import turtle as t
    
    t.colormode(255)
    t.pencolor((140,86,232))
    t.forward(100)
    t.right(90)
    t.pencolor(105,90,242)
    t.forward(200)
    t.right(90)
    t.pencolor(92,113,219)
    t.forward(100)
    t.right(90)
    t.pencolor(86,187,232)
    t.forward(200)
    t.right(90)
    t.penup()
    t.forward(6)
    t.right(90)
    t.forward(6)
    t.pendown()
    t.pencolor(230,143,69)
    t.forward(200)
    t.left(90)
    t.pencolor(240,172,72)
    t.forward(100)
    t.left(90)
    t.pencolor(218,175,76)
    t.forward(200)
    t.left(90)
    t.pencolor(240,206,72)
    t.forward(100)
    t.penup()
    t.forward(6)
    t.left(90)
    t.forward(196)
    t.pendown()
    t.pencolor(235,31,13)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.right(180)
    t.penup()
    t.forward(106)
    t.left(90)
    t.forward(198)
    t.pendown()
    t.pencolor(19,189,27)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.left(90)
    t.forward(4)
    t.done()

    
def secondsConverter():
    newWindow = Toplevel()
    newWindow.title('Seconds Coverter Program')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info = Label(newWindow, text="Convert a number of seconds into Days:Hours:Minutes:Seconds format.", fg="#82b843", bg="black", font="none 12")
    info.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
    instructions = Label(newWindow, text="Enter the number of seconds:", fg="white", bg="black")
    instructions.grid(row=1, column=0, sticky=E, padx=(10,0), pady=10)
    
    userInputBox = Entry(newWindow, width=25)
    userInputBox.grid(row=1, column=1, sticky=W, padx=(0,10), pady=10)

    
    def calculate():
        
        timeInSeconds = float(userInputBox.get())
    
        
        daysToReport = timeInSeconds//86400
        # There are 3600 seconds in 1 hour
        hoursToReport = (timeInSeconds%86400)//3600
        minutesToReport = ((timeInSeconds%86400)%3600)//60
        secondsToReport = ((timeInSeconds%86400)%3600)%60
        
        answer = str(int(daysToReport))  +"days : "+ str(int(hoursToReport))  +"mins : "+ str(int(minutesToReport))  +"hrs : "+ str(int(secondsToReport)) + "secs"
        
        answerLabel = Label(newWindow, text=answer, width=30, fg="white", bg="black")
        answerLabel.grid(row=3, column=0, columnspan=2, padx=2, pady=10)
        
        userInputBox.delete(0,END)
    
    calculateButton = Button(newWindow, text="Calculate", command=calculate)
    calculateButton.grid(row=2, column=0, columnspan=2, padx=10, pady=10)
    

    quitAppButton = Button(newWindow, text="Exit Subprogram", command=newWindow.destroy, fg="black", bg="red").grid(row=5, column=1, padx=(0, 10), pady=(0, 10),sticky=E)
    
    
    
def tempConverter():
    newWindow = Toplevel()
    newWindow.title('Temperature Coverter Program')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info = Label(newWindow, text="Convert the temperature in degrees to Celsius or Fahrenheit", fg="#82b843", bg="black", font="none 12")
    info.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
    
    celFrame = LabelFrame(newWindow, text="Celsius to Fahrenheit", bg="#82b843")
    celFrame.grid(row=1, column=0, padx=5, pady=5)
    fahFrame = LabelFrame(newWindow, text="Fahrenheit to Celsius", bg="#82b843")
    fahFrame.grid(row=1, column=1, padx=5, pady=5)
    
    instructionsC = Label(celFrame, text="Enter the temperature in Celsius:", fg="black", bg="#82b843")
    instructionsC.grid(row=0, column=0, sticky=E, padx=(10,0), pady=10)
    cInputBox = Entry(celFrame, width=25)
    cInputBox.grid(row=0, column=1, sticky=W, padx=(0,10), pady=10)
    
    instructionsF = Label(fahFrame, text="Enter the temperature in Fahrenheit:", fg="black", bg="#82b843")
    instructionsF.grid(row=0, column=0, sticky=E, padx=(10,0), pady=10)
    fInputBox = Entry(fahFrame, width=25)
    fInputBox.grid(row=0, column=1, sticky=W, padx=(0,10), pady=10)

    def calculateC2F():
        
        tempC = float(cInputBox.get())
        tempF = round(tempC*1.8+32,2)
        
        answer1 = "The temperature in Fahrenheit is " + str(tempF)
                      
        answerLabel = Label(celFrame, text=answer1, width=30, fg="black", bg="#82b843")
        answerLabel.grid(row=3, column=0, columnspan=2, padx=2, pady=10)
        
        cInputBox.delete(0,END)
        
    def calculateF2C():
        
        tempF = float(fInputBox.get())
        tempC = round((tempF-32)/1.8,2)
    

        answer = "The temperature in Celsius is " + str(tempC)
        
        answerLabel = Label(fahFrame, text=answer, width=30, fg="black", bg="#82b843")
        answerLabel.grid(row=3, column=0, columnspan=2, padx=2, pady=10)
        
        fInputBox.delete(0,END)    
    
    calcC2FButton = Button(celFrame, text="To Fahrenheit", command=calculateC2F)
    calcC2FButton.grid(row=2, column=0, columnspan=2, padx=10, pady=10)
    
    calcF2CButton = Button(fahFrame, text="To Celsius", command=calculateF2C)
    calcF2CButton.grid(row=2, column=0, columnspan=2, padx=10, pady=10)
    

    quitAppButton = Button(newWindow, text="Exit Subprogram", command=newWindow.destroy, fg="black", bg="red").grid(row=5, column=1, padx=(0, 10), pady=(0, 10),sticky=E)
        
def barChart():
    """
    function to display barchart in new window
    """
    import matplotlib.pyplot as plt
    from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
    
    
    newWindow = Toplevel()
    newWindow.title('US Population Bar Chart')
    newWindow.iconbitmap('rvcc_icon.ico')
    
    fig = plt.Figure(figsize = (8, 8))
    bar = fig.add_subplot(111)
    # Create a list with the X coordinates of each bar's left edge
    left_edges = [0, 10, 20, 30, 40]
    
    # Create a list iwht the heights of each bar
    heights = [266.28, 282.16, 295.52, 309.32, 320.64]
    
    # Create a variable for the bar width
    bar_width = 5
    
    # Add a title
    bar.title.set_text('United States population')
    
    # Add labels to the axes
    bar.set_xlabel('years')
    bar.set_ylabel('population')
    
    # Customize the tick marks
    bar.set_xticks([0, 10, 20, 30, 40])
    bar.set_xticklabels(['1995', '2000', '2005', '2010', '2015'])
    bar.set_yticks([250, 275, 300, 325, 350])
    bar.set_yticklabels(['250mil', '275mil', '300mil', '325mil', '350mil'])
    
    # Build the bar chart
    bar.bar(left_edges, heights, bar_width, 
            color=('m', 'b', 'k', 'r', 'y'))
    
    canvas = FigureCanvasTkAgg(fig, master=newWindow)
    #canvas.draw()
    canvas.get_tk_widget().pack()
    # Display the bar chart
    #plt.show()


def rockPaperScissors():
    newWindow = Toplevel()
    newWindow.title('Rock-Paper-Scissors Game')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info = Label(newWindow, text="Select Rock, Paper, or Scissors to go against the computer!", fg="#82b843", bg="black", font="none 12")
    info.grid(row=0, column=0, columnspan=3, padx=10, pady=10)
    
    choice = StringVar()
    #choice.set('rock')
    Radiobutton(newWindow, text='Rock', variable=choice, value='rock', fg="white", bg="black", selectcolor="black").grid(row=1, column=0, sticky=E, padx=6, pady=2)
    Radiobutton(newWindow, text='Paper', variable=choice, value='paper', fg="white", bg="black", selectcolor="black").grid(row=1, column=1, padx=6, pady=2)
    Radiobutton(newWindow, text='Scissors', variable=choice, value='scissors', fg="white", bg="black", selectcolor="black").grid(row=1, column=2, sticky=W, padx=6, pady=2)
    
    
    def play():
        
        userHand = choice.get()
        computerHand = random.choice(['rock','paper','scissors'])
        
        displayString = "Your choice is " + userHand + ", Computer's choice is " + computerHand
        
        showHands = Label(newWindow, text=displayString, width=50, fg="white", bg="black")
        showHands.grid(row=3, column=0, columnspan=3, padx=2, pady=10)
    
        if userHand == computerHand:
            result=("You have TIED!")
            state=1
        elif userHand == 'rock':
            if computerHand == 'paper':
                result=("You have LOST!")
                state=2
            if computerHand == 'scissors':
                result=("You have WON!")
                state=3
        elif userHand == 'paper':
            if computerHand == 'rock':
                result=("You have WON!")
                state=3
            if computerHand == 'scissors':
                result=("You have LOST!")
                state=2
        elif userHand == 'scissors':
            if computerHand == 'paper':
                result=("You have WON!")
                state=3
            if computerHand == 'rock':
                result=("You have LOST!")
                state=2
        
        if state==1:
            resultLabel = Label(newWindow, text=result, width=30, fg="white", bg="black", font="none 12 bold")
        elif state==2:
            resultLabel = Label(newWindow, text=result, width=30, fg="red", bg="black", font="none 12 bold")
        elif state==3:
            resultLabel = Label(newWindow, text=result, width=30, fg="#82b843", bg="black", font="none 12 bold")    
        resultLabel.grid(row=4, column=0, columnspan=3, padx=2, pady=10)
        
    
    playButton = Button(newWindow, text="GO!", command=play)
    playButton.grid(row=2, column=0, columnspan=3, padx=10, pady=10)
    
    quitAppButton = Button(newWindow, text="End Game", command=newWindow.destroy, fg="black", bg="red").grid(row=5, column=2, padx=(0, 10), pady=(0, 10),sticky=E)

    
    
def chooseYourOwn():
    newWindow = Toplevel()
    newWindow.title('Choose Your Own Adventure')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    startFrame = LabelFrame(newWindow, text="Welcome", padx=50, pady=5, bg="#82b843")
    startFrame.grid(row=0, column=0, columnspan=2, padx=50, pady=5)
    
    sentence1Story="You are the eldest prince of a medieval kingdom."
    Label(startFrame, text=sentence1Story, bg="#82b843", font="none 12 italic").grid(row=0, column=0, columnspan=2)
    sentence2Story="You wake up in your bedchambers with a hooded figure perched on your window."
    Label(startFrame, text=sentence2Story, bg="#82b843", font="none 12 italic").grid(row=1, column=0, columnspan=2)
    sentence3Story="The figure asks you to go with him if you want to live."
    Label(startFrame, text=sentence3Story, bg="#82b843", font="none 12 italic").grid(row=2, column=0, columnspan=2)
    
    def goWithHim():
        
        button1.configure(state="disabled")
        button2["state"]="disabled"
        
        secondFrame = LabelFrame(newWindow, text="Go with him", padx=50, pady=5, bg="#82b843")
        secondFrame.grid(row=1, column=0, columnspan=2, padx=50, pady=5)
        
        sentence1Story="The figure introduces himself as the humble but fiercely loyal groundskeeper."
        Label(secondFrame, text=sentence1Story, bg="#82b843", font="none 12 italic").grid(row=0, column=0, columnspan=2)
        sentence2Story="He also explains that he has overheard a plot to assasinate you from within the royal court."
        Label(secondFrame, text=sentence2Story, bg="#82b843", font="none 12 italic").grid(row=1, column=0, columnspan=2)
        sentence3Story="The groundskeeper explains that he has a waiting party in the swamps"
        Label(secondFrame, text=sentence3Story, bg="#82b843", font="none 12 italic").grid(row=2, column=0, columnspan=2)
        sentence4Story="with enough supplies to allow you to start a new life."
        Label(secondFrame, text=sentence4Story, bg="#82b843", font="none 12 italic").grid(row=3, column=0, columnspan=2)
        
        def agreeToMeet():
            
            button3.configure(state="disabled")
            button4["state"]="disabled"
            
            ending1="Upon arriving in the swamps, you are murdered by the 'humble' groundskeep for your royal possessions."
            ending2="Perhaps someone of your stature shouldn't be so naive."
            Label(newWindow, text=ending1, fg="red", bg="black", font="none 12 italic").grid(row=2, column=0, columnspan=2, pady=(100,0))
            Label(newWindow, text=ending2, fg="red", bg="black", font="none 12 italic").grid(row=3, column=0, columnspan=2, pady=(0,50))
            
            btn_text.set("End Game")
            
        
        def confrontCourt():
            
            button3.configure(state="disabled")
            button4["state"]="disabled"
            
            ending1="An investigation is launched by intelligence minister and it was found"
            ending2="that the groundskeeper was in fact plotting several high crimes."
            ending3="Such is the life of a member of the royal family."
            Label(newWindow, text=ending1, fg="#82b843", bg="black", font="none 12 italic").grid(row=2, column=0, columnspan=2, pady=(100,0))
            Label(newWindow, text=ending2, fg="#82b843", bg="black", font="none 12 italic").grid(row=3, column=0, columnspan=2)
            Label(newWindow, text=ending3, fg="#82b843", bg="black", font="none 12 italic").grid(row=4, column=0, columnspan=2, pady=(0,50))
            
            btn_text.set("End Game")
        
        
        button3=Button(secondFrame, text="Agree to meet the waiting party", command=agreeToMeet)
        button4=Button(secondFrame, text="Confront the royal court", command=confrontCourt)
        button3.grid(row=4, column=0, padx=3, pady=5, sticky=E)
        button4.grid(row=4, column=1, padx=3, pady=5, sticky=W)
    
    
    def callTheGuards():
        
        button1.configure(state="disabled")
        button2["state"]="disabled"
        
        secondFrame = LabelFrame(newWindow, text="Call the royal guards", padx=50, pady=5, bg="#82b843")
        secondFrame.grid(row=1, column=0, columnspan=2, padx=50, pady=5)
        
        sentence1Story="Upon yelling for the guards to arrive, the shadowy figure slinks away swiftly."
        Label(secondFrame, text=sentence1Story, bg="#82b843", font="none 12 italic").grid(row=0, column=0, columnspan=2)
        sentence2Story="The floor guards arrive hurriedly and ask what the matter is."
        Label(secondFrame, text=sentence2Story, bg="#82b843", font="none 12 italic").grid(row=1, column=0, columnspan=2)
        
        def explainDetails():
            
            button5.configure(state="disabled")
            button6["state"]="disabled"
            
            finalFrame = LabelFrame(newWindow, text="Explain the details of the stranger", padx=50, pady=5, bg="#82b843")
            finalFrame.grid(row=2, column=0, columnspan=2, padx=50, pady=5)
            
            sentence1Story="You explain who you saw and try to describe him in details, in hopes that"
            Label(finalFrame, text=sentence1Story, bg="#82b843", font="none 12 italic").grid(row=0, column=0, columnspan=2)
            sentence2Story="your securtiy forces can quickly apprehend him."
            Label(finalFrame, text=sentence2Story, bg="#82b843", font="none 12 italic").grid(row=1, column=0, columnspan=2)
            sentence3Story="Unfortunately, the guard assigned to your floor is loyal to someone else in the court."
            Label(finalFrame, text=sentence3Story, bg="#82b843", font="none 12 italic").grid(row=2, column=0, columnspan=2)
            sentence4Story="Enough times passes and the infiltrator is never caught."
            Label(finalFrame, text=sentence4Story, bg="#82b843", font="none 12 italic").grid(row=3, column=0, columnspan=2)
            sentence5Story="Days later, you are meeting with a high ranking regional commander in their camp tent over mulled wine."
            Label(finalFrame, text=sentence5Story, bg="#82b843", font="none 12 italic").grid(row=4, column=0, columnspan=2)
            sentence6Story="Never seeing him in a favorable manner, and after attempts over your life,"
            Label(finalFrame, text=sentence6Story, bg="#82b843", font="none 12 italic").grid(row=5, column=0, columnspan=2)
            sentence7Story="you decide whether you should take a sip."
            Label(finalFrame, text=sentence7Story, bg="#82b843", font="none 12 italic").grid(row=6, column=0, columnspan=2)
            
            def takeSip():
                
                button7.configure(state="disabled")
                button8["state"]="disabled"
                
                ending1="You take a cautious sip and continue with the meeting. It tastes delicious."
                ending2="Lowering your guard eventually, you have finished the whole cup."
                ending3="It is now apparent that the attempt on your life is successful,"
                ending4="as your chest tightens and your vision blurs."
                Label(newWindow, text=ending1, fg="red", bg="black", font="none 12 italic").grid(row=3, column=0, columnspan=2, pady=(100,0))
                Label(newWindow, text=ending2, fg="red", bg="black", font="none 12 italic").grid(row=4, column=0, columnspan=2)
                Label(newWindow, text=ending3, fg="red", bg="black", font="none 12 italic").grid(row=5, column=0, columnspan=2)
                Label(newWindow, text=ending4, fg="red", bg="black", font="none 12 italic").grid(row=6, column=0, columnspan=2, pady=(0,50))
                
                btn_text.set("End Game")
            
            def servantSip():
                
                button7.configure(state="disabled")
                button8["state"]="disabled"
                
                ending1="The commander makes a scene. No sip is ever taken by any party, but your wrest over"
                ending2="your kingdom has weakened and the citizens have cast doubt over you."
                ending3="Your duties as the royal heir have become more troublesome, but at least you are still in charge"
                ending4="of your own destiny as the future ruler of the kingdom, you think to yourself."
                Label(newWindow, text=ending1, fg="#82b843", bg="black", font="none 12 italic").grid(row=3, column=0, columnspan=2, pady=(100,0))
                Label(newWindow, text=ending2, fg="#82b843", bg="black", font="none 12 italic").grid(row=4, column=0, columnspan=2)
                Label(newWindow, text=ending3, fg="#82b843", bg="black", font="none 12 italic").grid(row=5, column=0, columnspan=2)
                Label(newWindow, text=ending4, fg="#82b843", bg="black", font="none 12 italic").grid(row=6, column=0, columnspan=2, pady=(0,50))
                
                btn_text.set("End Game")
            
            button7=Button(finalFrame, text="Take a sip", command=takeSip)
            button8=Button(finalFrame, text="Insult him and ask a servant to take a sip", command=servantSip)
            button7.grid(row=7, column=0, padx=3, pady=5, sticky=E)
            button8.grid(row=7, column=1, padx=3, pady=5, sticky=W)
        
        def fortifyCastle():
            
            button5.configure(state="disabled")
            button6["state"]="disabled"
            
            ending1="The royal army has soon barricaded all the walls, towers, and entrances and a search is done throughout."
            ending2="The shadowy figure is caught and it has turned out that he was a sworn servant of"
            ending3="the castle gone rogue, perhaps due to greed or even other unknown actors."
            ending4="You think to yourself that your grip on your castle needs to be tighter from now on."
            Label(newWindow, text=ending1, fg="#82b843", bg="black", font="none 12 italic").grid(row=3, column=0, columnspan=2, pady=(100,0))
            Label(newWindow, text=ending2, fg="#82b843", bg="black", font="none 12 italic").grid(row=4, column=0, columnspan=2)
            Label(newWindow, text=ending3, fg="#82b843", bg="black", font="none 12 italic").grid(row=5, column=0, columnspan=2)
            Label(newWindow, text=ending4, fg="#82b843", bg="black", font="none 12 italic").grid(row=6, column=0, columnspan=2, pady=(0,50))
            
            btn_text.set("End Game")
        
        
        button5=Button(secondFrame, text="Explain the details of the hooded figure", command=explainDetails)
        button6=Button(secondFrame, text="Immediately fortify the castle", command=fortifyCastle)
        button5.grid(row=3, column=0, padx=3, pady=5, sticky=E)
        button6.grid(row=3, column=1, padx=3, pady=5, sticky=W)
    
    
    button1=Button(startFrame, text="Go with him", command=goWithHim)
    button2=Button(startFrame, text="Call the royal guards", command=callTheGuards)
    button1.grid(row=3, column=0, padx=3, pady=5, sticky=E)
    button2.grid(row=3, column=1, padx=3, pady=5, sticky=W)
    
    
    btn_text = StringVar()
    quitAppButton = Button(newWindow, textvariable=btn_text, command=newWindow.destroy, fg="black", bg="red")
    btn_text.set("Quit early")
    quitAppButton.grid(row=7, column=1, padx=(0, 10), pady=(0, 10),sticky=E)

    
def coinFlipCasino():
    
    newWindow = Toplevel()
    newWindow.title('Coinflip Casino!')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info0 = Label(newWindow, text="Welcome, flip a coin and try your luck!", fg="#82b843", bg="black", font="none 12")
    info0.grid(row=0, column=0, columnspan=2, padx=10, pady=10)
    info1 = Label(newWindow, text="If heads, win $10", fg="white", bg="black", font="none 10")
    info1.grid(row=1, column=0, columnspan=2, padx=10, sticky=W)
    info2 = Label(newWindow, text="If tails, lose $10", fg="white", bg="black", font="none 10")
    info2.grid(row=2, column=0, columnspan=2, padx=10, pady=(1,20), sticky=W)
    
    inputMoney = Label(newWindow, text="How much do you want to play with? ($)", fg="white", bg="black").grid(row=3, column=0, sticky=E, padx=(10,0), pady=5)
    moneyInputBox = Entry(newWindow, width=25)
    moneyInputBox.insert(0,"0")
    moneyInputBox.grid(row=3, column=1, sticky=W, padx=(0,10), pady=5)
    
    inputTimes = Label(newWindow, text="How many times do you want to flip the coin?", fg="white", bg="black").grid(row=4, column=0, sticky=E, padx=(10,0), pady=5)
    timesInputBox = Entry(newWindow, width=25)
    timesInputBox.insert(0,"0")
    timesInputBox.grid(row=4, column=1, sticky=W, padx=(0,10), pady=5)
    
    headsCountLabel = Label(newWindow, text="Number of heads: 0", fg="white", bg="black")
    headsCountLabel.grid(row=6, column=0, sticky=W, padx=10)
    tailsCountLabel = Label(newWindow, text="Number of tails: 0", fg="white", bg="black")
    tailsCountLabel.grid(row=7, column=0, sticky=W, padx=10)
    
    def flipCoin():
        
        wallet = int(moneyInputBox.get())
        headsCount = 0
        tailsCount = 0
        
        timesToFlip = int(timesInputBox.get())
        
        for i in range(0,timesToFlip):
            coin = random.randrange(1,3)
            if coin == 1:
                
                headsCount += 1
                wallet += 10
            else:
                
                tailsCount += 1
                wallet -= 10
                
        headsCountLabel['text']="Number of heads: "+str(headsCount)
        tailsCountLabel['text']="Number of tails: "+str(tailsCount)
        
        cashOut = "You now have $" + str(wallet) + ".00 in your wallet."
        Label(newWindow, text=cashOut, width=30, fg="#82b843", bg="black", font="none 14").grid(row=8, column=0, columnspan=2, padx=5, pady=5)
        
    
    Button(newWindow, text="Flip coin", command=flipCoin).grid(row=5, column=1, pady=(5,20))

    quitAppButton = Button(newWindow, text="Leave Casino", command=newWindow.destroy, fg="black", bg="red").grid(row=9, column=1, padx=(0, 10), pady=(0, 10),sticky=E)
    
    
def contactManager():
    import csv
    
    newWindow = Toplevel()
    newWindow.title('Contact Manager Program')
    newWindow.configure(background="black")
    newWindow.iconbitmap('rvcc_icon.ico')
    
    info0 = Label(newWindow, text="Welcome to Contact Manager", fg="#82b843", bg="black", font="none 12")
    info0.grid(row=0, column=0, columnspan=2, padx=10, pady=(10,5))
    
    
    contactList = []
    
    try:
        with open("contacts.csv","r", newline="") as file:
            reader = csv.reader(file)
            for row in reader:
                contactList.append(row)
    except IOError:
        print("Contacts file not found.")
    
    global listFrame
    global viewFrame
    global addFrame
    global delFrame
    listFrame=LabelFrame(newWindow, text="Display all contacts", padx=50, pady=5, bg="#82b843")
    viewFrame=LabelFrame(newWindow, text="View a contact", padx=50, pady=5, bg="#82b843")
    addFrame=LabelFrame(newWindow, text="Add a contact", padx=50, pady=5, bg="#82b843")
    delFrame=LabelFrame(newWindow, text="Delete a contact", padx=50, pady=5, bg="#82b843")
    
    def displayContacts():
        """
        this function will print out the names in the current contact list
        """
        global listFrame
#         global viewFrame
#         global addFrame
#         global delFrame
        
        listFrame.destroy()
        viewFrame.destroy()
        addFrame.destroy()
        delFrame.destroy()
    
        listFrame=LabelFrame(newWindow, text="Display all contacts", padx=50, pady=5, bg="#82b843")
        listFrame.grid(row=2, column=0, columnspan=2, padx=20, pady=(5,20))

        if len(contactList) == 0:
            print("There are no contacts in the manager currently.")
            Label(listFrame, text="There are no contacts in the manager currently.", fg="red", bg="#82b843").grid()
        else:
            i = 1
            for contact in contactList:
                print(str(i) + ". " + contact[0])
                Label(listFrame, text=str(i) + ". " + contact[0], fg="black", bg="#82b843").grid(row=i, column=0, sticky=W)
                i += 1  
    
    def viewContact():
        """
        this function will display all details for a contact the user specifies
        """
#         global listFrame
        global viewFrame
#         global addFrame
#         global delFrame
        
        listFrame.destroy()
        viewFrame.destroy()
        addFrame.destroy()
        delFrame.destroy()
        
        viewFrame=LabelFrame(newWindow, text="View a contact", padx=50, pady=5, bg="#82b843")
        viewFrame.grid(row=2, column=0, columnspan=2, padx=20, pady=(5,20))
        
        
        if len(contactList) == 0:
            print("There are no contacts in the manager currently.")
            Label(viewFrame, text="There are no contacts in the manager currently.", fg="red", bg="#82b843").grid()
        else:
            
            Label(viewFrame, text="Number:", fg="black", bg="#82b843").grid(row=0, column=0, sticky=E)
            viewNumInput = Entry(viewFrame, width=5)
            viewNumInput.grid(row=0, column=1, sticky=W)
            
            def view():
                global viewFrame
                
                try:
                    contactNum = int(viewNumInput.get())
                    if contactNum < 1:
                        raise IndexError
                    print("Name: " + contactList[contactNum -1][0])
                    nameMsg = Label(viewFrame, text="Name: " + contactList[contactNum -1][0], fg="black", bg="#82b843")
                    nameMsg.grid(row=2, column=0, columnspan=2, sticky=W)
                    print("Email: " + contactList[contactNum -1][1])
                    emailMsg = Label(viewFrame, text="Email: " + contactList[contactNum -1][1], fg="black", bg="#82b843")
                    emailMsg.grid(row=3, column=0, columnspan=2, sticky=W)
                    print("Phone: " + contactList[contactNum -1][2])
                    phoneMsg = Label(viewFrame, text="Phone: " + contactList[contactNum -1][2], fg="black", bg="#82b843")
                    phoneMsg.grid(row=4, column=0, columnspan=2, sticky=W)
                    
                except IndexError:
                    print("Invalid contact number. Please try again.")
                except ValueError:
                    print("Please enter a contact number to view.")
                
                viewButton.destroy()
            
            
            viewButton = Button(viewFrame, text="View", command=view)
            viewButton.grid(row=1, column=0, columnspan=2, padx=5, pady=5)
    
    
    def addContact():
        """
        this function creates a contact with details specified by the user
        then that contact will get added to the end of the contact list in the manager
        """
#         global listFrame
#         global viewFrame
        global addFrame
#         global delFrame
        
        listFrame.destroy()
        viewFrame.destroy()
        addFrame.destroy()
        delFrame.destroy()
    
        addFrame=LabelFrame(newWindow, text="Add a contact", padx=50, pady=5, bg="#82b843")
        addFrame.grid(row=2, column=0, columnspan=2, padx=20, pady=(5,20))
    
        Label(addFrame, text="Name:", fg="black", bg="#82b843").grid(row=0, column=0, sticky=E)
        newNameInput = Entry(addFrame)
        newNameInput.grid(row=0, column=1, sticky=W)
        Label(addFrame, text="Email:", fg="black", bg="#82b843").grid(row=1, column=0, sticky=E)
        newEmailInput = Entry(addFrame)
        newEmailInput.grid(row=1, column=1, sticky=W)
        Label(addFrame, text="Phone:", fg="black", bg="#82b843").grid(row=2, column=0, sticky=E)
        newPhoneInput = Entry(addFrame)
        newPhoneInput.grid(row=2, column=1, sticky=W)
        
        def add():
            global addFrame
            
            name = newNameInput.get()
            email = newEmailInput.get()
            phone = newPhoneInput.get()
            newContact = []
            newContact.append(name)
            newContact.append(email)
            newContact.append(phone)
            contactList.append(newContact)
            print(name + " was added.")
            addedMsg = Label(addFrame, text=name + " was added.", fg="black", bg="#82b843")
            addedMsg.grid(row=4, column=0, columnspan=2)
            
            addButton.destroy()
    
        addButton = Button(addFrame, text="Add", command=add)
        addButton.grid(row=3, column=0, columnspan=2, padx=5, pady=5)
    
    def deleteContact():
        """
        this function will delete a contact specified by the user from the manager list
        """
#         global listFrame
#         global viewFrame
#         global addFrame
        global delFrame
        
        listFrame.destroy()
        viewFrame.destroy()
        addFrame.destroy()
        delFrame.destroy()
        
        delFrame=LabelFrame(newWindow, text="Delete a contact", padx=50, pady=5, bg="#82b843")
        delFrame.grid(row=2, column=0, columnspan=2, padx=20, pady=(5,20))
        
        
        if len(contactList) == 0:
            print("There are no contacts in the manager currently.")
            Label(delFrame, text="There are no contacts in the manager currently.", fg="red", bg="#82b843").grid()
        else:
            
            Label(delFrame, text="Number:", fg="black", bg="#82b843").grid(row=0, column=0, sticky=E)
            delNumInput = Entry(delFrame, width=5)
            delNumInput.grid(row=0, column=1, sticky=W)
            
            def delete():
                
                global delFrame
                
                try:
                    contactNum = int(delNumInput.get())
                    if contactNum < 1:
                        raise IndexError
                    
                    removedContact = contactList.pop(contactNum-1)
                    print(removedContact[0] + " was deleted.")
                    delMsg = Label(delFrame, text=removedContact[0] + " was deleted.", fg="black", bg="#82b843")
                    delMsg.grid(row=1, column=0, columnspan=2)
                    
                except IndexError:
                    print("Invalid contact number. Please try again.")
                except ValueError:
                    print("Please enter a contact number to view.")
                
                delButton.destroy()
            
            delButton = Button(delFrame, text="Delete", command=delete)
            delButton.grid(row=2, column=0, columnspan=2, padx=5, pady=5)
    
    
    commandFrame = LabelFrame(newWindow, text="Command Menu", padx=50, pady=5, bg="#82b843")
    commandFrame.grid(row=1, column=0, columnspan=2, padx=50, pady=(5,15))
    
    listButton = Button(commandFrame, text="Display all contacts", width=20, command=displayContacts)
    listButton.grid(row=0, column=0, padx=5, pady=5)
    viewButton = Button(commandFrame, text="View a contact", width=20, command=viewContact)
    viewButton.grid(row=0, column=1, padx=5, pady=5)
    addButton = Button(commandFrame, text="Add a contact", width=20, command=addContact)
    addButton.grid(row=1, column=0, padx=5, pady=5)
    delButton = Button(commandFrame, text="Delete a contact", width=20, command=deleteContact)
    delButton.grid(row=1, column=1, padx=5, pady=5)
    
    
    def saveAndExit():
        
        try:
            with open("contacts.csv","w", newline="") as file:
                writer = csv.writer(file)
                writer.writerows(contactList)
        except IOError:
            print("Contacts file not found.")
            
        newWindow.destroy()
    
    quitAppButton = Button(commandFrame, text="Save & Exit", command=saveAndExit, fg="black", bg="red").grid(row=2, column=0, columnspan=2, padx=(10, 10), pady=(10, 10))
    
    
    
    
labsFrame = LabelFrame(root, text="Semester Labs", padx=50, pady=5, bg="#82b843")
labsFrame.grid(row=3, column=0, columnspan=3, padx=50, pady=5)
lab2prob2button = Button(labsFrame,text="Lab2:Problem2", command=bitcoinConverter)
lab2prob3button = Button(labsFrame,text="Lab2:Problem3", command=nonsensicalStory)
lab3prob2button = Button(labsFrame,text="Lab3:Problem2", command=rockPaperScissors)
lab3prob3button = Button(labsFrame,text="Lab3:Problem3", command=chooseYourOwn)
lab4prob3button = Button(labsFrame,text="Lab4:Problem3", command=coinFlipCasino)
lab7prob1button = Button(labsFrame,text="Lab7:Problem1", command=contactManager)
lab2prob2button.grid(row=0, column=0, padx=5, pady=5)
lab2prob3button.grid(row=0, column=1, padx=5, pady=5)
lab3prob2button.grid(row=1, column=0, padx=5, pady=5)
lab3prob3button.grid(row=1, column=1, padx=5, pady=5)
lab4prob3button.grid(row=2, column=0, padx=5, pady=5)
lab7prob1button.grid(row=2, column=1, padx=5, pady=5)

hwFrame = LabelFrame(root, text="Semester Homeworks", padx=50, pady=5, bg="#82b843")
hwFrame.grid(row=4, column=0, columnspan=3, padx=50, pady=5)
hw2prob1button = Button(hwFrame,text="Hw2:Problem1", command=secondsConverter)
hw2prob2button = Button(hwFrame,text="Hw2:Problem2", command=tempConverter)
hw5prob1button = Button(hwFrame,text="Hw5:Problem1", command=artwork)
hw7prob2button = Button(hwFrame,text="Hw7:Problem2", command=barChart)
hw2prob1button.grid(row=0, column=0, padx=5, pady=5)
hw2prob2button.grid(row=0, column=1, padx=5, pady=5)
hw5prob1button.grid(row=1, column=0, padx=5, pady=5)
hw7prob2button.grid(row=1, column=1, padx=5, pady=5)


quitAppButton = Button(root, text="Exit Program", command=root.destroy, fg="black", bg="red").grid(row=5, column=1, padx=(0, 10), pady=(0, 10),sticky=E)
root.mainloop()
```
